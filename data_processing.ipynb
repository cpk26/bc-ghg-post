```python
import pandas as pd
import json
```

```python
#Read manually preprocessed sites file
site_data = pd.read_excel('sites.xlsx')
```

```python
#Remove rows without lat/long
site_data.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

```python
#Sort by descending size for leaflet plotting
site_data.sort_values('Grand Total', ascending=False, inplace=True, kind='quicksort')
```

```python
#Split the Primary Activity NAICS Code & Description column into two
new = site_data["Primary Activity NAICS Code & Description"].str.split("--", n = 1, expand = True) 
  
#NAICS Code column 
site_data["NAICS Code"]= new[0] 
  
#Primary Activity Column
site_data["Primary Activity"]= new[1] 

#Drop combined column
site_data.drop(columns =["Primary Activity NAICS Code & Description"], inplace = True) 

#Replace NaN with zeros for plotting
site_data.fillna(0, inplace=True)
```

```python
#List Unique NAICS CODE
print(site_data["NAICS Code"].unique())

#Drop Facility Type since only IF_a has lat/long
site_data.drop(columns =["Facility Type"], inplace = True) 
```

```python
#Export to json
site_data_json = site_data.to_json(path_or_buf='sites.json', orient='records')
```
